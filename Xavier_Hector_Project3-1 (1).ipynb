#### Bonus 

##### We can convert the equation $x^2-6=0$ into a fixed-point problem: $$x=x+c(x^2-6)$$ Find a $c$ such that the fixed point methods converges to the solution.

**The method of fix**: You want to find g(x)=x, you perform $$x_{n+1}=g(x_n).$$
You can use this method to find roots as well. Assume you want f(x)=0, you define g(x)=x+f(x) then the fixed point of g is the root of f $$g(x)=x \Longleftrightarrow x+f(x)=x \Longleftrightarrow f(x)=0$$


```python
# Define the function g(x)
def g(x, c):
    return x + c * (x**2 - 6)

# Fixed-point iteration with history
def fixed_point(c, x0=2, tol=1e-6, max_iter=100):
    x = x0
    history = [x]  # Step 1: store the initial guess
    for i in range(max_iter):
        x_new = g(x, c)  # Step 2: apply iteration formula
        history.append(x_new)  # Step 3: store new approximation
        if abs(x_new - x) < tol:  # Step 4: stop if close enough
            break
        x = x_new
    return x_new, history  # return both final value and all steps

# Choose one c
c = -0.1
root, history = fixed_point(c)

# True root
x_true = np.sqrt(6)

# Print the result
print(f"Using c = {c}")
print(f"Approximation: {root:.6f}")
print(f"Actual Root (√6): {x_true:.6f}")
print(f"Error: {abs(root - x_true):.6e}\n")
```
