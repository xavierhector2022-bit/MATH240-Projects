## Exercise 1

Plot the Taylor approximation of $\frac{1}{3x-x^2}$ of degree $4$ centered at $1$, for $x$ in $[1,3]$

```python
# Define function
def f(x):
    return 1 / (3*x - x**2)

# 4th-degree Taylor polynomial centered at a = 1
def P4(x):
    return (1/2
            - (1/4)*(x-1)
            + (3/8)*(x-1)**2
            - (5/16)*(x-1)**3
            + (11/32)*(x-1)**4)
```

```python
# Generate x-values
x = np.linspace(1, 2.5, 200)

# Compute function and approximation
y_true = f(x)
y_approx = P4(x)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y_true, label='f(x) = 1/(3x - x²)', linewidth=2)
plt.plot(x, y_approx, '--', label='P₄(x) centered at 1', linewidth=2)
plt.axvline(1, color='gray', linestyle=':', label='center (a=1)')
plt.title('4th-degree Taylor Approximation of f(x) = 1/(3x - x²)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

## Exercise 2

Plot the Taylor approximation of $\frac{1}{1-x}$ of degree $4$ centered at 0, for $x$ in $[-0.5,0.5]$

```python
# Define function
def f(x):
    return 1 / (1 - x)

# 4th-degree Taylor polynomial centered at a = 0
def P4(x):
    return 1 + x + x**2 + x**3 + x**4
```

```python
# Generate x-values
x = np.linspace(-0.5, 0.5, 200)

# Compute actual values and approximation
y_true = f(x)
y_approx = P4(x)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y_true, label='f(x) = 1/(1-x)', linewidth=2)
plt.plot(x, y_approx, '--', label='P₄(x) = 1 + x + x² + x³ + x⁴', linewidth=2)
plt.axvline(0, color='gray', linestyle=':', label='center (a=0)')
plt.title('4th-degree Taylor Approximation of f(x) = 1/(1-x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

## Exercise 3

1. Find and plot on the $x-$ interval $[\pi,3.3]$: $\ln(1+2(x-\pi))$, $P_1(x)$, $P_2(x)$, $P_3(x), P_4(x)$, where the Taylor polynomials are centered at $\pi$.
2. Which one is more accurate when you get close to $\pi$? What happens when you get away from $\pi$?

```python
# Function
def f(x):
    return np.log(1 + 2*(x - np.pi))

# Taylor polynomials centered at a=pi from the 1st-degree to the 4th one centered at a = pi
def P1(x): 
    return 2*(x - np.pi)
def P2(x): 
    return 2*(x - np.pi) - 2*(x - np.pi)**2
def P3(x):
    return 2*(x - np.pi) - 2*(x - np.pi)**2 + (8/3)*(x - np.pi)**3
def P4(x): 
    return 2*(x - np.pi) - 2*(x - np.pi)**2 + (8/3)*(x - np.pi)**3 - 4*(x - np.pi)**4
```

```python
# Generate x-values
x = np.linspace(np.pi, 3.3, 300)

# Plot
plt.figure(figsize=(9,6))
plt.plot(x, f(x), 'k', linewidth=2, label='f(x) = ln(1 + 2(x - π))')
plt.plot(x, P1(x), '--', label='P₁(x)')
plt.plot(x, P2(x), '--', label='P₂(x)')
plt.plot(x, P3(x), '--', label='P₃(x)')
plt.plot(x, P4(x), '--', label='P₄(x)')
plt.axvline(np.pi, color='gray', linestyle=':', label='center (π)')
plt.title('Taylor Polynomials of ln(1 + 2(x - π)) around x = π')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

##### **P4 is the most accurate as we get closer to $\pi$. As you get away from $\pi$, the accuracy drops**

## Exercise 4
1. Find and plot on the $x-$ interval $[-1,1]$, the quadratic Taylor polynomials of $e^{-\cos(x)}$ centered at $x=0$.
2. Plot the error between the Taylor of polynomial of degree 2 for $e^{-\cos(x)}$, centered at 0. 
3. Bound the error in this approximation when $-1\leq x \leq 1$. ($Error(x)=|Approximation(x) - Exact(x)|$)

```python
# Define function and its 2nd-degree Taylor polynomial centered at a = 0
def f(x):
    return np.exp(-np.cos(x))

def P2(x):
    return np.exp(-1) + (np.exp(-1)/2)*x**2
```

```python
# Generate x-values
x = np.linspace(-1, 1, 400)

# Compute true and approximate values
y_true = f(x)
y_approx = P2(x)
error = np.abs(y_true - y_approx)

# Plot function and polynomial
plt.figure(figsize=(8,5))
plt.plot(x, y_true, label='f(x) = e^{-cos(x)}', linewidth=2)
plt.plot(x, y_approx, '--', label='P₂(x)', linewidth=2)
plt.axvline(0, color='gray', linestyle=':', label='center (a=0)')
plt.title('Quadratic Taylor Approximation of $e^{-\\cos(x)}$ around x=0')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot error
plt.figure(figsize=(8,5))
plt.plot(x, error, color='red', linewidth=2)
plt.title('Error |f(x) - P₂(x)| on [-1, 1]')
plt.xlabel('x')
plt.ylabel('Error')
plt.grid(True)
plt.show()

# Maximum error value
print("Maximum error on [-1, 1]:", np.max(error))
```

## Exercise 6
A well known approximation, when $x$ is small, is $\cos(x)\simeq 1-x^2/2$. 
1. Bound the error of the approximation of $\sin(x)$ by $x$ for $-\pi/4\leq x \leq \pi/4$ using the plot. This means the maximum of $|\cos(x)-(1-x^2/2)|$ when $x$ is $-\pi/4\leq x \leq \pi/4$.
2. Bound the error using the formula.
3. Compare.

```python
# The function
def f(x):
    return np.cos(x)

# The 2nd-degree Taylor polynomial centered at a = 0
def P2(x):
    return 1 - (x**2) / 2

# Define interval
x_values = np.linspace(-np.pi/4, np.pi/4, 400)

# Compute values
E_vals = f(x_values) - P2(x_values)

# Find the maximum absolute error and corresponding x
E_max = np.max(abs(E_vals))

# Bounding the Error using the plot 
print(f"Maximum |E(x)| ≈ {(E_max):.8f}")

# Compute the actual error bound from Taylor's remainder formula
taylor_remainder = (1/24) * (np.pi/4)**4
print(f"|R2(x)| = |f(x) - P2(x)| ≤ {taylor_remainder:.8f}")
```

```python
# Plot error with bounds
plt.figure(figsize=(8,5))
plt.plot(x_values, E_vals, label=r'$E(x) = \cos(x) - (1 - \frac{x^2}{2})$', color='blue')
plt.axhline(taylor_remainder, color='green', linestyle='--', label= "Taylor remainder bound")
plt.title(r'Error of $\cos(x) \approx 1 - \frac{x^2}{2}$ on $[-\pi/4, \pi/4]$')
plt.xlabel('x')
plt.ylabel('Error E(x)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Define g(x)
def g(x):
    return (np.exp(x) - 1) / x

# Taylor polynomials of e^x from 1st-degree up to the third centered at a = 0
def P1(x): 
    return 1 + x
def P2(x): 
    return 1 + x + x**2/2
def P3(x): 
    return 1 + x + x**2/2 + x**3/6

# Corresponding approximations for g(x) using the Taylor polinomial of e^x 
def g1(x): 
    return (P1(x) - 1) / x
def g2(x): 
    return (P2(x) - 1) / x
def g3(x): 
    return (P3(x) - 1) / x
```

## Bonus 
Using Taylor approximation of degree 3 approximate $\int_0^\pi e^{\cos(x)}dx$.

```python
# Define the original function
def f(x):
    return np.exp(np.cos(x))

# The 3rd-degree Taylor polynomial centered at a = 0
def P3(x):
    return np.e * (1 - (x**2)/2)

# Integration limits
x_vals = np.linspace(0, np.pi, 100)

# Integration using trapezoidal rule
I_true = np.trapezoid(f(x_vals), x_vals)
I_approx = np.trapezoid(P3(x_vals), x_vals)

print(f"Approximation (Taylor degree 3): {I_approx:.4f}")
print(f"True integral: {I_true:.4f}")
print(f"Error: {abs(I_true - I_approx):.4f}")
```
